### **1-3. Simple Linear Regression**

- Scikit-Learn ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÏÇ¨Ïö©ÌïòÏó¨ Îã®ÏàúÏÑ†ÌòïÌöåÍ∑Ä Î™®Îç∏ÏùÑ ÎßåÎìúÏÑ∏Ïöî.
  - ÌäπÏÑ±ÏùÄ `sqft_living`Ïù¥Í≥†, ÌÉÄÍ≤üÏùÄ `price`ÏûÖÎãàÎã§.
- Î™®Îç∏ ÌïôÏäµ ÌõÑ, ÌäπÏÑ± `sqft_living`Ïóê ÎåÄÌïú ÌÉÄÍ≤ü `price`Ïùò ÏòàÏ∏°Í∞íÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. 

```python
from sklearn.linear_model import LinearRegression

feature = ['sqft_living']
target = 'price'

x_sim = df[feature]
y = df[target]

print(x_sim.shape)
print(y.shape)
```

```python
simple_ols = LinearRegression()
simple_ols.fit(x_sim, y)
```

##### **Q3. Îã®ÏàúÏÑ†ÌòïÌöåÍ∑Ä Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú Í≤∞Í≥ºÏóêÏÑú indexÍ∞Ä 0Ïù∏ Í∞íÏùÑ Ï†úÏ∂úÌï¥ Ï£ºÏÑ∏Ïöî. (ÏÜåÏàòÏ†ê Ïù¥Ìïò Î∞òÏò¨Î¶º)**

```python
## Ïó¨Í∏∞ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥Ï£ºÏÑ∏Ïöî
y_sim = simple_ols.predict(x_sim)
round(y_sim[0])
```

### **1-4. Multiple Linear Regression**

- Scikit-Learn ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÏÇ¨Ïö©ÌïòÏó¨ Îã§Ï§ëÏÑ†ÌòïÌöåÍ∑Ä Î™®Îç∏ÏùÑ ÎßåÎìúÏÑ∏Ïöî. 
  - ÌäπÏÑ±ÏùÄ `sqft_living`, `bathrooms`Ïù¥Í≥†, ÌÉÄÍ≤üÏùÄ `price`ÏûÖÎãàÎã§.
- Î™®Îç∏ ÌïôÏäµ ÌõÑ, Îëê ÌäπÏÑ± `sqft_living`, `bathrooms`Ïóê ÎåÄÌïú ÌÉÄÍ≤ü `price`Ïùò ÏòàÏ∏°Í∞íÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. 

```python
features = ['sqft_living', 'bathrooms']

x_mul = df[features]

print(x_mul.shape)
print(y.shape)
```

```python
multiple_ols = LinearRegression()
multiple_ols.fit(x_mul, y)
```

##### **Q4. Îã§Ï§ëÏÑ†ÌòïÌöåÍ∑Ä Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú Í≤∞Í≥ºÏóêÏÑú indexÍ∞Ä 0Ïù∏ Í∞íÏùÑ Ï†úÏ∂úÌï¥ Ï£ºÏÑ∏Ïöî. (ÏÜåÏàòÏ†ê Ïù¥Ìïò Î∞òÏò¨Î¶º)**

```python
## Ïó¨Í∏∞ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥Ï£ºÏÑ∏Ïöî
y_mul = multiple_ols.predict(x_mul)
round(y_mul[0])
```

##### **Q5. Îã§Ï§ëÏÑ†ÌòïÌöåÍ∑Ä Î™®Îç∏Ïùò $R^2$ÏùÑ Î∞òÏò¨Î¶ºÌïòÏó¨ ÏÜåÏàò Ï≤´Ïß∏ ÏûêÎ¶¨ÍπåÏßÄ Íµ¨Ìïú Í∞íÏùÑ Ï†úÏ∂úÌï¥ Ï£ºÏÑ∏Ïöî.**

```python
## Ïó¨Í∏∞ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥Ï£ºÏÑ∏Ïöî
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

multiple_r2 = r2_score(y, y_mul)
multiple_r2.round(1)
```

### **4-4. ÌèâÍ∞ÄÏßÄÌëú**

**4-2**, **4-3**ÏóêÏÑú Íµ¨Ìïú Îã®ÏàúÏÑ†ÌòïÌöåÍ∑Ä, Îã§Ï§ëÏÑ†ÌòïÌöåÍ∑ÄÏùò ÏÑ±Îä•ÏùÑ ÌèâÍ∞ÄÌï¥ Î¥ÖÎãàÎã§.
- Í∞Å Î™®Îç∏Ïùò mse, rmse, mae, $R^2$ Í∞íÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. 
- ÏúÑÏóêÏÑú Íµ¨Ìïú ÌèâÍ∞ÄÏßÄÌëúÏôÄ **Part 1**ÏóêÏÑú Íµ¨Ìïú ÌèâÍ∞ÄÏßÄÌëú Ïä§ÏΩîÏñ¥Ïóê ÎåÄÌïòÏó¨, Ìï¥Îãπ Í≤∞Í≥ºÎ•º ÌÜµÌï¥ ÏñªÏùÑ Ïàò ÏûàÎäî Ïù∏ÏÇ¨Ïù¥Ìä∏Îäî Ïñ¥Îñ§ Í≤ÉÏù¥ ÏûàÎäîÏßÄ ÏÑ§Î™ÖÌï¥ Î≥¥ÏÑ∏Ïöî.

```python
y_real = df['price']

def eval_models(y_pred, y_real=y_real) :
    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_real, y_pred)
    r2 = r2_score(y_real, y_pred)

    return mse, rmse, mae, r2
```

```python
simple2_mse, simple2_rmse, simple2_mae, simple2_r2 = eval_models(y_sim2)
multiple2_mse, multiple2_rmse, multiple2_mae, multiple2_r2 = eval_models(y_mul2)
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)

comparison_metrics = pd.DataFrame(index=['mse', 'rmse', 'mae', 'r2'], columns=['Simple', 'Multiple'])
comparison_metrics['Simple'] = [simple2_mse, simple2_rmse, simple2_mae, simple2_r2]
comparison_metrics['Multiple'] = [multiple2_mse, multiple2_rmse, multiple2_mae, multiple2_r2]
comparison_metrics
```
