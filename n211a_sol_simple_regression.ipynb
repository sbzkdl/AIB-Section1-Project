### **1-3. Simple Linear Regression**

- Scikit-Learn 라이브러리를 사용하여 단순선형회귀 모델을 만드세요.
  - 특성은 `sqft_living`이고, 타겟은 `price`입니다.
- 모델 학습 후, 특성 `sqft_living`에 대한 타겟 `price`의 예측값을 구하세요. 

```python
from sklearn.linear_model import LinearRegression

feature = ['sqft_living']
target = 'price'

x_sim = df[feature]
y = df[target]

print(x_sim.shape)
print(y.shape)
```

```python
simple_ols = LinearRegression()
simple_ols.fit(x_sim, y)
```

##### **Q3. 단순선형회귀 모델이 예측한 결과에서 index가 0인 값을 제출해 주세요. (소수점 이하 반올림)**

```python
## 여기에서 과제를 진행해주세요
y_sim = simple_ols.predict(x_sim)
round(y_sim[0])
```

### **1-4. Multiple Linear Regression**

- Scikit-Learn 라이브러리를 사용하여 다중선형회귀 모델을 만드세요. 
  - 특성은 `sqft_living`, `bathrooms`이고, 타겟은 `price`입니다.
- 모델 학습 후, 두 특성 `sqft_living`, `bathrooms`에 대한 타겟 `price`의 예측값을 구하세요. 

```python
features = ['sqft_living', 'bathrooms']

x_mul = df[features]

print(x_mul.shape)
print(y.shape)
```

```python
multiple_ols = LinearRegression()
multiple_ols.fit(x_mul, y)
```

##### **Q4. 다중선형회귀 모델이 예측한 결과에서 index가 0인 값을 제출해 주세요. (소수점 이하 반올림)**

```python
## 여기에서 과제를 진행해주세요
y_mul = multiple_ols.predict(x_mul)
round(y_mul[0])
```

### **4-4. 평가지표**

**4-2**, **4-3**에서 구한 단순선형회귀, 다중선형회귀의 성능을 평가해 봅니다.
- 각 모델의 mse, rmse, mae, $R^2$ 값을 구하세요. 
- 위에서 구한 평가지표와 **Part 1**에서 구한 평가지표 스코어에 대하여, 해당 결과를 통해 얻을 수 있는 인사이트는 어떤 것이 있는지 설명해 보세요.

```python
y_real = df['price']

def eval_models(y_pred, y_real=y_real) :
    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_real, y_pred)
    r2 = r2_score(y_real, y_pred)

    return mse, rmse, mae, r2
```

```python
simple2_mse, simple2_rmse, simple2_mae, simple2_r2 = eval_models(y_sim2)
multiple2_mse, multiple2_rmse, multiple2_mae, multiple2_r2 = eval_models(y_mul2)
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)

comparison_metrics = pd.DataFrame(index=['mse', 'rmse', 'mae', 'r2'], columns=['Simple', 'Multiple'])
comparison_metrics['Simple'] = [simple2_mse, simple2_rmse, simple2_mae, simple2_r2]
comparison_metrics['Multiple'] = [multiple2_mse, multiple2_rmse, multiple2_mae, multiple2_r2]
comparison_metrics
```
